```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn import metrics
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
```

```python
startups_data = pd.read_csv('/content/50_Startups.csv')
```

```python
startups_data.tail()
```

```python
# Checking for some information about dataset
startups_data.info()
```

Thus, there are no missing values

DATA VISUALIZATION

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 3, 1)
sns.scatterplot(data=startups_data, x='R&D Spend', y='Profit')
plt.title('R&D Spend vs Profit')

plt.subplot(1, 3, 2)
sns.scatterplot(data=startups_data, x='Administration', y='Profit')
plt.title('Administration vs Profit')

plt.subplot(1, 3, 3)
sns.scatterplot(data=startups_data, x='Marketing Spend', y='Profit')
plt.title('Marketing Spend vs Profit')

plt.tight_layout()
plt.show()
```

Thus, R&D spend and marketing spend have a nearly perfect linear relationship with Profit generated.

```python
sns.pairplot(startups_data)
plt.show
```

```python
# Correlation Heatmap
plt.figure(figsize=(10, 8))
correlation_matrix = startups_data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix')
plt.show()
```

Thus, there is a negative correlation between administrative expenses and Marketing Spend whereas there's a very high correlation between R&D spend and Profit.

```python
# Box Plot
plt.figure(figsize=(14, 6))
plt.subplot(1, 3, 1)
sns.boxplot(data=startups_data, x='R&D Spend')
plt.title('R&D Spend Distribution')

plt.subplot(1, 3, 2)
sns.boxplot(data=startups_data, x='Administration')
plt.title('Administration Distribution')

plt.subplot(1, 3, 3)
sns.boxplot(data=startups_data, x='Marketing Spend')
plt.title('Marketing Spend Distribution')

plt.tight_layout()
plt.show()
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
# Checking the distribution of each column, in case of outliers
plt.figure(figsize=(14, 8))

plt.subplot(2, 2, 1)
sns.distplot(startups_data['R&D Spend'], kde=True, bins=20)
plt.title('Distribution of R&D Spend')

plt.subplot(2, 2, 2)
sns.distplot(startups_data['Administration'], kde=True, bins=20)
plt.title('Distribution of Administration')

plt.subplot(2, 2, 3)
sns.distplot(startups_data['Marketing Spend'], kde=True, bins=20)
plt.title('Distribution of Marketing Spend')

plt.subplot(2, 2, 4)
sns.distplot(startups_data['Profit'], kde=True, bins=20)
plt.title('Distribution of Profit')

plt.tight_layout()
plt.show()
```

Thus, most of them do not have any major outliers.

```python
# Separating data and labels
X = startups_data.drop(columns = 'Profit', axis = 1)
Y = startups_data['Profit']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

MODEL TRAINING USING LINEAR REGRESSION

```python
model_linear = LinearRegression()
```

```python
model_linear.fit(X_train, Y_train)
```

```python
training_data_prediction_linear = model_linear.predict(X_train)
```

```python
# R squared error
error_score_linear = metrics.r2_score(Y_train, training_data_prediction_linear)
print(error_score_linear)
```

```python
testing_data_prediction_linear = model_linear.predict(X_test)
```

```python
error_score_testing_data_linear = metrics.r2_score(Y_test, testing_data_prediction_linear)
print(error_score_testing_data_linear)
```

```python
plt.scatter(Y_test, testing_data_prediction_linear)
plt.xlabel("Actual price")
plt.ylabel("Predicted price")
plt.title("Actual price vs predicted price")
plt.show()
```

MODEL TRAINING USING LASSO REGRESSION

```python
model_lasso = Lasso()
```

```python
model_lasso.fit(X_train, Y_train)
```

```python
training_data_prediction_lasso = model_lasso.predict(X_train)
```

```python
error_score_lasso = metrics.r2_score(Y_train, training_data_prediction_lasso)
print(error_score_lasso)
```

```python
testing_data_prediction_lasso = model_lasso.predict(X_test)
```

```python
error_score_lasso_2 = metrics.r2_score(Y_test, testing_data_prediction_lasso)
print(error_score_lasso_2)
```

```python
plt.scatter(Y_test, testing_data_prediction_lasso)
plt.xlabel("Actual price")
plt.ylabel("Predicted price")
plt.title("Actual price vs predicted price")
plt.show()
```

MODEL TRAINING WITH RANDOM FOREST REGRESSOR

```python
model_regressor = RandomForestRegressor(n_estimators = 100)
```

```python
model_regressor.fit(X_train, Y_train)
```

```python
test_data_prediction_regressor = model_regressor.predict(X_test)
```

```python
print(test_data_prediction_regressor)
```

```python
error_score_regressor = metrics.r2_score(Y_test, test_data_prediction_regressor)
print(error_score_regressor)
```

```python
Y_test = list(Y_test)
```

```python
plt.plot(Y_test, color = 'Blue', label = 'Actual values')
plt.plot(test_data_prediction_regressor, color = 'Red', label = 'Predicted values')
plt.title('Actual vs predicted price')
plt.xlabel('Actual price')
plt.ylabel('Predicted price')
plt.legend()
plt.show()
```

MODEL TRAINING WITH XGBREGRESSOR

```python
model_XGB = XGBRegressor()
```

```python
model_XGB.fit(X_train, Y_train)
```

```python
training_data_prediction_XGB = model_XGB.predict(X_train)
```

```python
# R squared value
r2_train_xgb = metrics.r2_score(Y_train, training_data_prediction_XGB)
print("R squared value on training data:", r2_train_xgb)
```

```python
testing_data_prediction_XGB = model_XGB.predict(X_test)
```

```python
r2_test_xgb = metrics.r2_score(Y_test, testing_data_prediction_XGB)
print("R squared value on testing data:", r2_test_xgb)
```

BUILDING A PREDICTIVE SYSTEM BASED ON ALL 4 MODELS

The output should come as: 152211.77

```python
# Building a predictive system based on Linear Regression
input_data_linear = (120542.52, 148718.95, 311613.29)
input_data_as_numpy_array_linear = np.asarray(input_data_linear)

input_data_reshaped_linear = input_data_as_numpy_array_linear.reshape(1,-1)
prediction_linear = model_linear.predict(input_data_reshaped_linear)

print("The predicted profit according to linear regression is:", prediction_linear)
```

```python
# Building a predictive system based on Lasso Regression
input_data_lasso = (120542.52, 148718.95, 311613.29)
input_data_as_numpy_array_lasso = np.asarray(input_data_lasso)

input_data_reshaped_lasso = input_data_as_numpy_array_lasso.reshape(1,-1)
prediction_lasso = model_lasso.predict(input_data_reshaped_lasso)

print("The predicted profit according to Lasso regression is:", prediction_lasso)
```

```python
# Building a predictive system based on RandomForestRegressor
input_data_regressor = (120542.52, 148718.95, 311613.29)
input_data_as_numpy_array_regressor = np.asarray(input_data_regressor)

input_data_reshaped_regressor = input_data_as_numpy_array_regressor.reshape(1,-1)
prediction_regressor = model_regressor.predict(input_data_reshaped_regressor)

print("The predicted profit according to RandomForestRegressor is:", prediction_regressor)
```

```python
# Building a predictive system based on XGBoostRegressor
input_data_xgb = (120542.52, 148718.95, 311613.29)
input_data_as_numpy_array_xgb = np.asarray(input_data_xgb)

input_data_reshaped_xgb = input_data_as_numpy_array_xgb.reshape(1,-1)
prediction_xgb = model_XGB.predict(input_data_reshaped_xgb)

print("The predicted profit according to XGBoost Regressor is:", prediction_xgb)
```

Thus, the most reliable model amongst the 4 is XGBoost Regressor.
